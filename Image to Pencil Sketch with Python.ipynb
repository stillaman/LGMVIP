Image to Pencil Sketch with Python 

```python
# Importing Libraries

#!pip install opencv-python
import cv2 as cv
from IPython.display import Image
```

#### Loading the Image 

```python
Image("picture.jpeg", width = 900, height = 600)
```

```python
image = cv.imread("picture.jpeg")
```

#### Converting Image into gray scaled image mode

```python
gray_image = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
```

#### Inverting the gray scaled image

```python
inverted_image = 255 - gray_image
```

#### Blurring the gray scaled image using GaussianBlur Function

```python
blurred = cv.GaussianBlur(inverted_image, (21, 21), 0)
```

#### Inverting Blurred Image to Pencil Image

```python
inverted_blurred = 255 - blurred
pencil_sketch = cv.divide(gray_image, inverted_blurred, scale=256.0)
```

```python
cv.imshow("Original Landscape Image", image)
cv.imshow("Pencil Made Landscape Sketch", pencil_sketch)
cv.waitKey(0)
```
