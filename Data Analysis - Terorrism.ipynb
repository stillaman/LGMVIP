```python
# Importing required Libraries
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Loading Dataset
df=pd.read_csv('globalterrorismdb_0718dist.csv')
df.head()
```

```python
# Features of Dataset
print(df.columns)
```

```python
# Heatmap of Missing Value
sns.heatmap(df.isnull())
plt.show()
```

```python
#Value count of Region
df.region_txt.value_counts()
```

```python
# Cases during the Years
plt.figure(figsize=(6,6))
sns.displot(df.iyear,bins=20)
plt.title('Cases during the years')
plt.xlabel('Years 1970 to 2010')
plt.ylabel('Count of cases')
plt.show()
```

```python
# Region_Txt Vs Region
plt.figure(figsize=(10,6))
groups=df.groupby(['region_txt'])['region'].sum()
groups.plot.bar(color='red')
plt.xlabel('Region_txt')
plt.ylabel('Region')
plt.show()
```

#### From above graph we found that most of the terrorism events  happened in Middle East and North Africa followed by South Asia and Sub-Saharan Africa.

```python
# Attack type 1 as per Region and Country
Attack_region=pd.pivot_table(data=df,
                             values='attacktype1',
                             index=['region_txt','country_txt'],
                             aggfunc=[np.sum,np.mean],
                             fill_value=0)
print(Attack_region)
```

```python
# Attacks in Australasia & Oceania - Total and Average 
Attack_region.query('region_txt==["Australasia & Oceania"]')
```

```python
# Visuals of Attacks in Australasia & Oceania - Total and Average 
sns.histplot(Attack_region.query('region_txt==["Australasia & Oceania"]'))
plt.title('Visuals of Attacks in Australasia & Oceania - Total and Average ')
plt.show()
```

```python
# Attacks in South Asia - Total and Average 
Attack_region.query('region_txt==["South Asia"]')
```

```python
# Visuals of Attacks in South Asia - Total and Average 
sns.histplot(Attack_region.query('region_txt==["South Asia"]'))
plt.title('Visuals of Attacks in South Asia - Total and Average ')
plt.show()
```

```python
# Attacks in South Asian country India - Total and Average 
Attack_region.query('country_txt==["India"]')
```

```python
# Attacks in Middle East & North Africa - Total and Average 
Attack_region.query('region_txt==["Middle East & North Africa"]')
```

```python
# Different types of attack during the years -Attack type 1
sns.set_theme(style="ticks")
plt.figure(figsize=(20,6))
sns.histplot(data=df,
             x='iyear',
             bins=10,
             multiple="stack",
             #palette="light:m_r",
             edgecolor=".3",
             linewidth=.5,
             #log_scale=True,
             hue='attacktype1_txt'
            )
plt.title('Different types of Attack during the years - Attack Type 1')
plt.xlabel('Years 1970 to 2010')
plt.ylabel('Count of cases')
plt.show()
```

```python
# Different types of attack during the years - Attack type 2
sns.set_theme(style="ticks")
plt.figure(figsize=(20,6))
sns.histplot(data=df,
             x='iyear',
             bins=10,
             multiple="stack",
             #palette="light:m_r",
             edgecolor=".3",
             linewidth=.5,
             #log_scale=True,
             hue='attacktype2_txt'
            )
plt.title('Different types of Attack during the years - Attack Type 2')
plt.xlabel('Years 1970 to 2010')
plt.ylabel('Count of cases')
plt.show()
```

#### From the above plot we found that the maximum number or event of terrorist attacks happened in the year 2010, and in 2010 majorly all events happened due to Bombing/Explosion and next was due to Armed assualt.

```python
# Different types of attack during the years - Attack type 2
#sns.set_theme(style="ticks")
plt.figure(figsize=(10,6))
sns.displot(data=df,
             x='iyear',
             bins=10,
             #multiple="stack",
             #palette="light:m_r",
             #edgecolor=".3",
             #linewidth=.5,
             #log_scale=True,
             hue='targtype1_txt',
            element='step')

plt.title('Different types of Attack during the years - Target Type 1')
plt.xlabel('Years 1970 to 2010')
plt.ylabel('Count of cases')
plt.show()
```

#### From above plot we found that main targets were Private Citizens & property followed by Governement diplomatic, Journalists and Media and so on as we can infer from the plot or graph

#### In summary, most of the cases started escalating from 1970 however sudden escalation or spikes can be noticed during 2010 in Middle East and North Africa followed by South Asia and Sub-Saharan Africa and in most of the cases primary targets were Private Citizens & property followed by Governement diplomatic, Journalists and Media.
